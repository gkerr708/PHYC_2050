```python
import numpy as np, matplotlib.pyplot as plt
```

```python
def f(x,y):
    return x**2 + y**2
```

```python
x, y = np.linspace(-1,1,100), np.linspace(-1,1,100)
X, Y = np.meshgrid(x,y)  
Z = f(X,Y)
```

```python
fig = plt.figure(figsize=(8,8))
axes = plt.axes(projection='3d')
axes.plot_surface(X,Y,Z, cmap='cool', lw=0.5, rstride=1, cstride=1, alpha=0.5)
_=plt.show()
```
